```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre
from scipy.special import spherical_jn
from scipy.special import spherical_yn
from scipy.optimize import minimize
```

```python
A = np.linspace(16,250,1000)
k = .22*10**(15)
K = 1.54*10**(15)
kR = k*1.13*A
KR = K*1.13*A
delta = np.arctan((k*np.cos(kR)*np.sin(KR)-K*np.cos(KR)*np.sin(kR))/(k*np.sin(kR)*np.sin(KR)-K*np.cos(KR)*np.cos(kR)))
```

```python
sigma = (4*np.pi/k**2)*(np.sin(delta))**2
plt.plot(A,sigma)
```
