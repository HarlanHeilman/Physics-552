```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre
from scipy.special import spherical_jn
from scipy.special import spherical_yn
```

```python
k = 2*10**(15)
R = 1*10**(-10)
i = 1j

j0 = spherical_jn(0,R*k)
n0 = spherical_yn(0,R*k)
j1 = spherical_jn(1,R*k)
n1 = spherical_yn(1,R*k)
j2 = spherical_jn(2,R*k)
n2 = spherical_yn(2,R*k)
j3 = spherical_jn(3,R*k)
n3 = spherical_yn(3,R*k)
j4 = spherical_jn(4,R*k)
n4 = spherical_yn(4,R*k)

P0 = legendre(0)
P1 = legendre(1)
P2 = legendre(2)
P3 = legendre(2)

d0 = np.arctan(j0/n0)
d1 = np.arctan(j1/j2)
d2 = np.arctan(j2/n2)
d3 = np.arctan(j3/n3)
d4 = np.arctan(j4/n3)

a0 = (2*0+1)
a1 = (2*1+1)
a2 = (2*2+1)
a3 = (2*3+1)
a4 = (2*4+1)
```

```python
theta = np.linspace(0,np.pi,100)
mu = np.cos(theta)
f1 = (a0*(-1)/(2*i*k)*P0(mu))
f2 = (a1*(-1)/(2*i*k)*P1(mu))
F1 = (a0*(np.exp(i*np.pi)-1)/(2*i*k)*P0(mu))
F2 = (a1*(np.exp(i*np.pi)-1)/(2*i*k)*P1(mu))
#f3 = (1/k)*(a2*np.exp(d2*i)*np.sin(d2)*P2(mu))
#f4 = (1/k)*(a3*np.exp(d3*i)*np.sin(d3)*P2(mu))
#f5 = (1/k)*(a4*np.exp(d4*i)*np.sin(d4)*P2(mu))
#f4 = f3+(1/k)*(a3*np.exp(d3*i)*np.sin(d3)*P0(mu))
diff_cross = abs(f1+f2+f3)
diff_cross2 = abs(F1+F2)
```

```python
plt.plot(mu,diff_cross,label = r'$|f(\theta)|^2$')
plt.plot(mu,diff_cross2,label = r'$|F(\theta)|^2$')
#plt.plot(mu,diff_cross_0,label = r'$|f(\theta)|^2$')
#plt.plot(mu,diff_cross_1,label = r'$|f(\theta)|^2$')
plt.ylabel('Differential cross Section')
plt.xlabel(r'$\cos(\theta)$')
plt.legend()
```
